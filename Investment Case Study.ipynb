- We know the country and investment type now we need to decide the sector.We have multiple categories,So we use the mapping file to map them to one main sector out of 8.

```python
mapping=pd.read_csv('C:\\UpGrad\\mapping.csv')
```

```python
import re
```

```python
def removezeros(data):
    ret_val=[]
    if data==data:
        for i in data.split(' '):
            if re.match('[a-zA-Z]*0[a-zA-Z]*',i)!=None:
                ret_val.append(i.replace('0','na'))
            else:
                ret_val.append(i)
    return ' '.join(ret_val).title()
```

- Above function will remove the `0` present insted of `na` in category list of mapping file as data cleaning process and put `na` in place `0` using regex.
- Since we need to merge them,So we bring to comman case in this case title(camelcase)

```python
mapping['category_list']=mapping['category_list'].map(removezeros)
```

```python
master_frame['primary_sector']=master_frame['category_list'].map(lambda x:x.split('|')[0].title())
```

- Their are multiple categories so we take the first category as primary category by doing the above operation.

```python
master_frame.head()
```

```python
final_df=master_frame.merge(mapping,left_on='primary_sector',right_on='category_list',how='left')
```

- Merging the dataframes to form the dataframe with sectors mapped.

- ##### From the above output we can make out that their are some primary sectors which are not present in the mapping file.
- ##### So we map them to others sector as they are also one of the important sectors,So we cannot remove them from analysis.

```python
final_df.loc[final_df['category_list_y'].isnull(),['Automotive & Sports','Blanks','Cleantech / Semiconductors','Entertainment','Health','Manufacturing','News, Search and Messaging','Others','Social, Finance, Analytics, Advertising']]=[0,0,0,0,0,0,0,1,0]
```

- ##### We need to create common main sector instead of maintaining these many sectors in seperate columns.

```python
def test(row):
    if row['Automotive & Sports']==1.0:
        return 'Automotive & Sports'
    elif row['Cleantech / Semiconductors']==1.0:
        return 'Cleantech / Semiconductors'
    elif row['Entertainment']==1.0:
        return 'Entertainment'
    elif row['Health']==1.0:
        return 'Health'
    elif row['Manufacturing']==1.0:
        return 'Manufacturing'
    elif row['News, Search and Messaging']==1.0:
        return 'News, Search and Messaging'
    elif row['Others']==1.0:
        return 'Others'
    else:
        return 'Social, Finance, Analytics, Advertising'
    
```

- ##### Above function will determine the main sector along the row which will be added to Main Sector column.

```python
final_df['Main Sector']=final_df.apply(test,axis=1)
```
